# Identifiers

<p>
A Python identifier is the name used to identify variables, functions, classes, modules, and other objects in a Python program. Identifiers help programmers refer to different elements of the code in a meaningful way. They can contain letters (A-Z, a-z), digits (0-9), and underscores (_), but they cannot start with a digit. Python identifiers are case-sensitive, which means `name`, `Name`, and `NAME` are considered different identifiers. Special characters such as `@`, `#`, and `$` are not allowed in identifiers. Additionally, Python keywords like `if`, `for`, and `while` cannot be used as identifiers because they have predefined meanings in the language. Using clear and descriptive identifiers improves code readability and makes programs easier to understand and maintain.

# Statements 

## Single line Statement 

<p>
A single-line statement in Python is an instruction that is written and completed on a single line. It is the most common way of writing code and is used for simple operations such as variable assignment, printing output, or performing calculations. Single-line statements improve readability and make programs easy to understand. Examples include assigning a value to a variable (x = 10) or displaying text (print("Hello")). Each statement is executed line by line by the Python interpreter.
</p>

```python
x = 10 + 25 + 30 + 99
print(x)
```

## Multiline Statement

<p>
A multi-line statement is a statement that extends across two or more lines. It is used when a statement becomes too long to fit comfortably on one line, making the code more readable and organized. Python allows multi-line statements by using a backslash (\) for explicit line continuation or by placing the statement inside parentheses (), brackets [], or braces {} for implicit continuation. Multi-line statements are commonly used for long mathematical expressions, large lists, dictionaries, or complex function calls. They help keep code neat and easier to maintain.
</p>

```python
x = 12 + 34 + 45 + 67 + 44 \
+ 23 + 122 + 34 \
+ 34

print(x)
```

# Variables

<p>
A variable in Python is a named storage location used to hold data that can be accessed and modified during program execution. Variables allow programmers to store different types of information such as numbers, text, lists, and objects. In Python, a variable is created automatically when a value is assigned to it using the assignment operator (`=`), and there is no need to declare its data type explicitly. The value stored in a variable can be changed at any time, making variables flexible and useful for handling dynamic data. Meaningful variable names improve code readability and make programs easier to understand. Variables play a fundamental role in programming because they enable data manipulation, calculations, and information management throughout a program.

# DataTypes

<p>A datatype in Python defines the kind of value a variable can store and determines the operations that can be performed on that value. Python provides several built-in data types to handle different kinds of data efficiently. Common data types include integers (`int`) for whole numbers, floating-point numbers (`float`) for decimal values, strings (`str`) for text, booleans (`bool`) for logical values (`True` or `False`), lists (`list`) for ordered collections, tuples (`tuple`) for immutable collections, sets (`set`) for unique elements, and dictionaries (`dict`) for key-value pairs. Python is a dynamically typed language, meaning the data type of a variable is determined automatically when a value is assigned. Understanding data types is essential because they influence how data is stored, processed, and manipulated within a program.</p>


```python
name = "Taleeb"
age = 21
weight = 70.5
isStudent = True

print(type(name),name)
print(type(age),age)
print(type(weight),weight)
print(type(isStudent),isStudent)
```
